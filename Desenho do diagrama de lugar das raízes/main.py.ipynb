# Desenho diagrama lugar das raízes

O diagrama de lugar das raízes é uma forma de representarmos como os zeros e as raízes de um sistema podem variar de acordo com um parâmetro k que pode ser ajustado. O desenho desse tipo de diagrama envolve vários passos que estão descritos nos slides do Prof. Sérgio Pires Pimentel e no livro texto da disciplina. Mas no python, é bem fácil de fazer esse desenho. Considere a seguinte função de transferência de malha aberta:

$$FTMA = K \frac{S^2 - 4S + 20}{(S+2)(S+4)}$$

Vamos desenhar o seu lugar das raízes:

```python
!pip install control numpy matplotlib
```

```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

Primeiro, definimos a função de malha aberta para a biblioteca control

```python
FTMA = ctrl.tf([1,-4,20], [1, 6, 8])
print(FTMA)
```

Agora podemos desenhar o diagrama simplesmente chamando a função root_locus da biblioteca control e passando a FTMA. Se passarmos o parâmetro plot=True, a biblioteca irá procurar um elemento do matplotlib chamado figure no nosso código e desenhará o gráfico lá dentro. Ao contrário, se passarmos plot=False, a função irá nos retornar dois arrays, um com os valores da variação de k e outro com as raízes para aquele k definido. O parâmetro grid=True nos ajuda a desenhar linhas guias no nosso gráfico.

```python
plt.figure()
ctrl.root_locus(FTMA, plot=True, grid=True)
plt.title("Diagrama de lugar das raízes")
plt.xlabel("Parte Real")
plt.ylabel("Parte Imaginária")
plt.show()
```
